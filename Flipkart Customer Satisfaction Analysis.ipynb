## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# 1. Standardizing column names: lowercase and replacing spaces with underscores
customer_data.columns = customer_data.columns.str.lower().str.replace(' ', '_')

# 2. Drop connected_handling_time due to excessive missing values
customer_data.drop(columns=['connected_handling_time'], inplace=True)

# 3. Drop duplicate rows if any
customer_data.drop_duplicates(inplace=True)

# 4. Handle null 'Order_id' (Filled with 'Unknown')
customer_data['order_id'] = customer_data['order_id'].fillna('Unknown')

# 5. Filled missing categorical values with 'Unknown'
categorical_cols = ['customer_city', 'product_category']
customer_data[categorical_cols] = customer_data[categorical_cols].fillna('Unknown')

# 6. Filled missing customer_remarks values with 'no_feedback'
customer_data['customer_remarks'] = customer_data['customer_remarks'].fillna("no_feedback")

# 7. Converted date columns while handling different formats
def convert_date(date_str):
    formats = ["%d/%m/%Y %H:%M", "%d-%b-%y"]  # Possible formats
    for fmt in formats:
        try:
            return pd.to_datetime(date_str, format=fmt)
        except ValueError:
            continue
    return pd.NaT  

customer_data['issue_reported_at'] = customer_data['issue_reported_at'].apply(lambda x: convert_date(x) if pd.notna(x) else pd.NaT)
customer_data['issue_responded'] = customer_data['issue_responded'].apply(lambda x: convert_date(x) if pd.notna(x) else pd.NaT)
customer_data['survey_response_date'] = customer_data['survey_response_date'].apply(lambda x: convert_date(x) if pd.notna(x) else pd.NaT)

# 8. # Createed 'resolution_time' new column in minutes and also handled negative values
customer_data['resolution_time'] = (customer_data['issue_responded'] - customer_data['issue_reported_at']).dt.total_seconds() / 60
customer_data['resolution_time'] = customer_data['resolution_time'].abs()

# 9.Filled missing 'item_price' using median as per 'Product_category' price

category_median_price = customer_data.groupby('product_category')['item_price'].median()

customer_data['item_price'] = customer_data.apply(lambda row: category_median_price[row['product_category']] 
                            if pd.isna(row['item_price']) and row['product_category'] in category_median_price 
                            else row['item_price'], axis=1)


# 10. Filled 'order_date_time' using 'issue_reported_at' - median resolution time
median_resolution_time = customer_data['resolution_time'].median()  # Calculate median resolution time in hours
customer_data['order_date_time'] = customer_data.apply(lambda row: row['issue_reported_at'] - pd.Timedelta(hours=median_resolution_time) 
                                 if pd.isna(row['order_date_time']) and pd.notna(row['issue_reported_at']) 
                                 else row['order_date_time'], axis=1)





````

### What all manipulations have you done and insights you found?

1. Standardized Column Names – Changed all column names to lowercase and replaced spaces with underscores for consistency.


2. Dropped Unnecessary Columns – Removed connected_handling_time because it had too many missing values.


3. Removed Duplicates – Deleted repeated rows to keep the dataset clean.


4. Handled Missing Values

    Replaced missing Order ID with "Unknown".

    Filled missing Customer City and Product Category with "Unknown".
    
    Replaced empty Customer Remarks with "no_feedback".

5. Converted Dates to Proper Format – Fixed inconsistencies in issue_reported_at, issue_responded, and survey_response_date by converting them to date-time format.


6. Created Resolution_Time Column – Calculated how long it took to resolve an issue by subtracting issue reported time from issue responded time (converted negative values to positive).


7. Filled Missing Prices – Used the median price of each product category to fill missing Item Prices.


8. Estimated Missing Order Dates – Used median resolution time to estimate missing order_date_time.

#### Chart - 11 - Issues Reported vs Issues Responded Over Time

````python
# Chart - 11 visualization code

reported_dates = customer_data['issue_reported_at'].dt.date.value_counts().sort_index()
responded_dates = customer_data['issue_responded'].dt.date.value_counts().sort_index()

sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 8))

# Lineplot for issues reported
sns.lineplot(x=reported_dates.index,y=reported_dates.values,label="Issues Reported",color="blue",marker="o",linewidth=2)
sns.lineplot(x=responded_dates.index,y=responded_dates.values,label="Issues Responded",color="green",marker="o",linewidth=2)

plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(color='grey', linestyle='--', linewidth=0.5)

plt.xlabel("Date", fontsize=14, fontweight='bold')
plt.ylabel("Count", fontsize=14, fontweight='bold')
plt.title("Issues Reported vs Issues Responded Over Time", fontsize=16, fontweight='bold', color="darkred")


plt.legend(title="Legend", title_fontsize=12, fontsize=10, loc="upper left")


highest_reported_date = reported_dates.idxmax()
highest_reported_count = reported_dates.max()
plt.annotate(f"Peak: {highest_reported_count}",
             xy=(highest_reported_date, highest_reported_count),
             xytext=(highest_reported_date, highest_reported_count + 5),
             arrowprops=dict(facecolor='blue', arrowstyle="->"),
             fontsize=10, color='blue')

highest_responded_date = responded_dates.idxmax()
highest_responded_count = responded_dates.max()
plt.annotate(f"Peak: {highest_responded_count}",
             xy=(highest_responded_date, highest_responded_count),
             xytext=(highest_responded_date, highest_responded_count + 5),
             arrowprops=dict(facecolor='green', arrowstyle="->"),
             fontsize=10, color='green')
plt.tight_layout()
````
